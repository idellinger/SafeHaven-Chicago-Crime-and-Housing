```python
import pandas as pd
import numpy as np
import sqlite3
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import seaborn as sns
import joblib
from sklearn import *
```

```python
database_file = 'C:\Data Science\chicago-crime-property-analysis\Israel\data\chicago-crime-property.db'
connection = sqlite3.connect(database_file)
cursor = connection.cursor()

table_query = "SELECT * FROM property_with_crime"
```

```python
df = pd.read_sql_query(table_query, connection)

connection.close()
```

```python
df_head = df.head()
df_head.to_csv('df_head.csv', index=False)
df_head.head()
```

```python
# Assuming 'neighborhood' is the column with neighborhood names and 'crimes' is the column with crime counts
plt.figure(figsize=(15, 10))
sns.barplot(x='neighborhood', y='crime_count', data=df)
plt.xticks(rotation=90)
plt.title('Crime Distribution by Neighborhoods')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Crimes')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.scatterplot(x='crime_count', y='sold_price', data=df)


# Define a custom currency formatter function
def currency_formatter(x, pos):
    return '${:,.0f}'.format(x)

# Apply the currency formatter to the Y-axis
formatter = FuncFormatter(currency_formatter)

plt.gca().yaxis.set_major_formatter(formatter)
plt.title('Correlation between Crimes and Property Values')
plt.xlabel('Number of Crimes')
plt.ylabel('Property Values')
plt.show()
```

```python
df['last_sold_date'] = pd.to_datetime(df['last_sold_date'])
df_grouped = df.groupby(pd.Grouper(key='last_sold_date', freq='Q')).size().reset_index(name='count')
df_grouped.info()
```

```python
merged_df = pd.merge(df, df_grouped, on='last_sold_date', how='right')
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='last_sold_date', y='crime_count', data=merged_df, palette='viridis')
plt.title('Crime Trends by Quarter')
plt.xlabel('Quarter')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
numeric_df = df.select_dtypes(include='number')
correlation_matrix = numeric_df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap (Numeric Variables Only)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='year_built', y='crime_count', data=df, alpha=0.7)
plt.title('Relationship between Crime Count and Year Built')
plt.xlabel('Year Built')
plt.ylabel('Crime Count')
plt.grid(True)
plt.show()
```

```python
df['year_built'].value_counts()
```

```python
filtered_years = df['year_built'][df['year_built'] > 1615]

if not filtered_years.empty:
    lowest_value_above_minus1 = np.min(filtered_years)
    print(f"The lowest value above -1 in 'year_built' is: {lowest_value_above_minus1}")
else:
    print("No values above -1 in 'year_built'.")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=filtered_years, y='crime_count', data=df, alpha=0.7)
plt.title('Relationship between Crime Count and Year Built')
plt.xlabel('Year Built')
plt.ylabel('Crime Count')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.jointplot(x=filtered_years, y='crime_count', data=df, kind='hex', color='blue')
plt.suptitle('Hexbin Plot: Relationship between Crime Count and Year Built', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.jointplot(x=filtered_years, y='crime_count', data=df, kind='kde', color='blue')
plt.suptitle('KDE Plot: Relationship between Crime Count and Year Built', y=1.02)
plt.show()
```
