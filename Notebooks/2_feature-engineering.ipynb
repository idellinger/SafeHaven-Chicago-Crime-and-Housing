### Crime Rates Relative to Population

- Crime Rate per Capita by calculating the total crimes per capita for each region/community and crime type. 
- This can help normalize the crime figures and provide a more comparable metric across different areas, regardless of their population size.

```python
# calculate crime rates per 1,000 population
crime_df['Crime_Rate_per_1000'] = (crime_df['incident_count'] / crime_df['TOT_POP']) * 1000
```

```python
crime_df.tail(2)
```

```python
violent_crimes_df = crime_df[crime_df['crime_category'] == 'Violent Crimes']

# calculate crime rate per 1000 population for Violent Crimes
crime_df['Crime_Rate_per_1000_violent'] = (violent_crimes_df['incident_count'] / violent_crimes_df['TOT_POP']) * 1000

property_crimes_df = crime_df[crime_df['crime_category'] == 'Property Crimes']

# calculate crime rate per 1000 population for Property Crimes
crime_df['Crime_Rate_per_1000_property'] = (property_crimes_df['incident_count'] / property_crimes_df['TOT_POP']) * 1000

drug_related_crimes_df = crime_df[crime_df['crime_category'] == 'Drug-Related Crimes']

# calculate crime rate per 1000 population for Drug-related Crimes
crime_df['Crime_Rate_per_1000_drug'] = (drug_related_crimes_df['incident_count'] / drug_related_crimes_df['TOT_POP']) * 1000
```

## Economic and Employment Indicators

### Income Per Capita

Provides insight into the economic status of a community, which could be correlated with types of crimes.

```python
crime_df['Income_Per_Capita'] = (crime_df['MEDINC'] * 1000) / crime_df['TOT_POP']
```

```python
crime_df.head(2)
```

### Employment Rate

This could highlight the relationship between employment and crime rates.

```python
crime_df['Employment_Rate'] = crime_df['TOT_EMP_RES'] / crime_df['TOT_POP']
```

```python
crime_df.tail(4)
```

### Crime Density

This would help in understanding how crime clusters geographically.

```python
conversion_factor = 27878400

# converting square feet to square miles
crime_df['shape_area_sq_miles'] = crime_df['shape_area'] / conversion_factor
```

### Season and Temporal Features

```python
crime_df['date'] = pd.to_datetime(crime_df['date'])

crime_df['Month'] = crime_df['date'].dt.month
crime_df['Day_of_Week'] = crime_df['date'].dt.day_name()
```

```python
def get_season(month):
    if month in (3, 4, 5):
        return 'Spring'
    elif month in (6, 7, 8):
        return 'Summer'
    elif month in (9, 10, 11):
        return 'Fall'
    else:
        return 'Winter'

crime_df['Season'] = crime_df['Month'].apply(get_season)
```
